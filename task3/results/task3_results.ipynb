```python
import polars as pl
import polars.selectors as cs
```

```python
df_scipy_500 = pl.read_csv("results_mtp_scipy_bfgs_500.csv").drop(
    "success", "final_loss"
)
df_scipy_1000 = pl.read_csv("results_mtp_scipy_bfgs_1000.csv").drop(
    "success", "final_loss"
)
df_scipy_001 = pl.read_csv("results_mtp_scipy_bfgs_001.csv").drop(
    "success", "final_loss"
)

df_mlip_500 = pl.read_csv(
    "/Users/peterzarenkov/projects/course_work/optimization_methods_MLP/task1/results/results_mtp.csv"
)

def display_ensemble_stats(df: pl.DataFrame, name: str):
    metrics = df.select(cs.numeric())
    display(pl.DataFrame({"stat": [name]}).hstack(metrics.select(pl.all().mean())))
    display(pl.DataFrame({"stat": [f"{name}_std"]}).hstack(metrics.select(pl.all().std())))
```

Срванение встроенного BFGS и scipy на 500 итерациях

```python
display(df_scipy_500)
display(df_mlip_500)
```

Среднее значение ошибки и разброс для scipy BFGS и встроенного.

```python
display_ensemble_stats(df_scipy_500, "scipy_500_mean")
display_ensemble_stats(df_mlip_500, "mlip_500_mean")
```

Сравниение ошибок scip BFGS для 500 итераций и 1000

```python
display(df_scipy_500)
display(df_scipy_1000)
```

Видим, что для некоторых потенциалов из ансамбля обучение останавливалось раньше и ошибки упали.

```python
display_ensemble_stats(df_scipy_500, "scipy_500_mean")
display_ensemble_stats(df_scipy_1000, "scipy_1000_mean")
```

Сравниение scipy BFGS на начальном интервале 0.01 и встроенный на том же интервале.

```python
display(df_scipy_001)
display(df_mlip_500)
```

```python
display_ensemble_stats(df_scipy_001, "scipy_001_mean")
display_ensemble_stats(df_mlip_500, "mlip_500_mean")
```

Здесь видим уже значительное преимущество scipy
