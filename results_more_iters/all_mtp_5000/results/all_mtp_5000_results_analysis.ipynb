## Краткие выводы

- Native BFGS даёт лучший validation EPA RMSE среди MTP-запусков.
- SciPy BFGS близок к native BFGS по energy error, но хуже по force error и не имеет успешных завершений по флагу SciPy в этой серии.
- Среди mini-batch методов лучшим по validation EPA RMSE остаётся Adam.
- Momentum лучше Muon и SGD по validation EPA RMSE, но уступает Adam и BFGS-вариантам.
- Muon pad sqrt лучше Muon factorization по validation EPA RMSE и forces RMSE, но не превосходит Adam.
