```python
pot_json = b''

rb = RadialBasisCinf(size = 8, min_dist = 2.0, cutoff = 5.0)
#pot = MTP(rb, [0], level = 8)
pot = NewPairPot(rb, [0])
pot.params[:] = np.random.uniform(low=-0.01, high=0.01, size=len(pot.params))
print(len(pot.params))
print(pot.params)

pot_json = pot.to_json_bytes()
func.attach_pot(pot)
grads = np.zeros(len(pot.params))
func.accumulate_grads(grads)
print(func.calc())
print(grads)
```

```python
tau = 1e-9

#Gradient descent
for i in range(0,100):
    pot.params[:] -= tau*grads[:]
    print(func.calc())
    grads = np.zeros(len(pot.params))
    func.accumulate_grads(grads)
```
